```stata
// 
// Contains data from /Users/jiangheng/Desktop/Work/Courses/Management Methods 2/term paper/_filed/overtakes/ALL_full_overtakes_season.dta
// Observations:         6,038
// Variables:            28                  27 Mar 2026 16:30
// ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
// Variable      Storage   Display    Value
// name         type    format    label      Variable label
// ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
// Year            long    %12.0g
// Round           long    %12.0g
// EventName       str27   %27s
// LapNumber       double  %10.0g
// Overtaker       str3    %3s
// OvertakerNumber long    %12.0g
// OvertakerTeam   str17   %17s
// PositionBefore  double  %10.0g
// PositionAfter   double  %10.0g
// Overtaken       str3    %3s
// OvertakenNumber long    %12.0g
// OvertakenTeam   str17   %17s
// SessionTime     str22   %22s
// SameLapWindow~t str22   %22s
// SameLapWindow~d str22   %22s
// PitInfluenceF~g byte    %8.0g
// DNFInfluenceF~g byte    %8.0g
// IsTeammate      byte    %8.0g
// Is_DRS_Active   str5    %5s
// RelativeDista~e double  %10.0g
// Throttle_Drop~x double  %10.0g
// Brake_Pressur~a double  %10.0g
// TeamOrderConf~e double  %10.0g
// Vmin_Deviation  double  %10.0g
// OffTrack_Flag   byte    %8.0g
// Lockup_Flag     byte    %8.0g
// LocalYellowFlag byte    %8.0g
// OffTrackConfi~e double  %10.0g
// ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
// Sorted by:
// 

// Count overtakes between each pair of drivers

// Normalize pair so that A vs B and B vs A are treated as the same pair
gen driver1 = Overtaker
gen driver2 = Overtaken

// Sort drivers alphabetically within each pair
replace driver1 = Overtaken if Overtaker > Overtaken
replace driver2 = Overtaker if Overtaker > Overtaken

gen pair = driver1 + " vs " + driver2

// Count frequency of each pair
bysort pair: gen pair_count = _N
bysort pair: keep if _n == 1

// Keep only top 10 pairs
gsort -pair_count
keep if _n <= 10

// Plot
graph hbar pair_count, over(pair, sort(pair_count) descending label(angle(0))) ///
    title("Top 10 Driver Pairs by Number of Overtakes") ///
    ytitle("Number of Overtakes") ///
    blabel(bar) ///
    scheme(s2color)
```
